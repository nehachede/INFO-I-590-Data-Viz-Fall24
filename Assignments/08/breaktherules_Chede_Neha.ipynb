```python
import pandas as pd
import numpy as np
import vega_datasets
import matplotlib.pyplot as plt
```

```python
iris = vega_datasets.data.iris()
iris.head()
```

```python
petal_length = iris['petalLength'].values

# histogram
plt.figure(figsize=(10, 5))

# default
plt.subplot(1, 2, 1)
plt.hist(petal_length, alpha=0.7)
plt.title("Histogram with Default Method")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

# Sturges' method
plt.subplot(1, 2, 2)
plt.hist(petal_length, bins='sturges', alpha=0.7)
plt.title("Histogram using Sturges' Method")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
```

```python
petal_length.size
```

```python
# add random data to petal lengths
new_petal_length = sum_array=np.asarray(np.concatenate((petal_length, np.random.normal(size = 1000))))
new_petal_length.size
```

```python
# histogram
plt.figure(figsize=(10, 5))

# default
plt.subplot(1, 2, 1)
plt.hist(new_petal_length, alpha=0.7)
plt.title("Histogram with Default Method")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")

# Sturges' method
plt.subplot(1, 2, 2)
plt.hist(new_petal_length, bins='sturges', alpha=0.7)
plt.title("Histogram using Sturges' Method")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
```

```python
plt.figure(figsize=(10, 10))

# Default method
plt.subplot(2, 2, 1)
plt.hist(new_petal_length, alpha=0.7)
plt.title("Histogram with Default Method")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Sturges' method
plt.subplot(2, 2, 2)
plt.hist(new_petal_length, bins='sturges', alpha=0.7)
plt.title("Histogram using Sturges' Method")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Rice method
plt.subplot(2, 2, 3)
plt.hist(new_petal_length, bins='rice', alpha=0.7)
plt.title("Histogram using Rice Method")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Square root method
plt.subplot(2, 2, 4)
plt.hist(new_petal_length, bins='sqrt', alpha=0.7)
plt.title("Histogram using Square Root Method")
plt.xlabel("Value")
plt.ylabel("Frequency")
```
